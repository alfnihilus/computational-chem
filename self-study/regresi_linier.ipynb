```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# @markdown ## Data from the problem

# @markdown ## Please enter concentration & absorbance data
x_data ="" # @param {type:"string"}
y_data ="" # @param {type:"string"}

concentration = np.array([float(x.strip()) for x in x_data.split(',')]).reshape(-1, 1)
absorbance = np.array([float(x.strip()) for x in y_data.split(',')])

# Perform Linear Regression
model = LinearRegression()
model.fit(concentration, absorbance)

# Regression parameters
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(concentration, absorbance)

# y = slope * x + intercept  => x = (y - intercept) / slope
sample_absorbance_str = "0" # @param {type:"string"}
sample_absorbance = float(sample_absorbance_str)
sample_concentration = (sample_absorbance - intercept) / slope

print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
print(f"R-squared: {r_squared}")
print(f"Sample Concentration (at A={sample_absorbance_str}): {sample_concentration}")

# Create plot
plt.scatter(concentration, absorbance, color='blue', label='Data Standar')
plt.plot(concentration, model.predict(concentration), color='red', label=f'y = {slope:.4f}x + {intercept:.4f} (R^2 = {r_squared:.3f})')
# If you don’t want to input sample data, you can turn it into a comment
# plt.scatter([sample_concentration], [sample_absorbance], color='green', marker='x', s=100, label='Sampel ' f"A = {sample_absorbance_str}")
x_axis = "" # @param {type:"string"}
plt.xlabel(x_axis)
y_axis = "" # @param {type:"string"}
plt.ylabel(y_axis)
title = "" # @param {type:"string"}
plt.title(title)
plt.legend()
plt.grid(False) # Changed from True to False to remove grid lines
plt.savefig('kurva_kalibrasi.png')
```
